```python
import pandas as pd
import matplotlib.pyplot as plt
import os
from google.colab import drive

drive.mount('/content/drive')
MODULE_PATH = "/content/drive/MyDrive/2025CV"
CHECKPOINT_DIR = os.path.join(MODULE_PATH, "checkpoint_jw_content")
```

```python
backbone_name = "EfficientNet-B3"  # 다른 백본이면 이름만 바꿔줘
history_file = os.path.join(CHECKPOINT_DIR, f"{backbone_name}_history.csv")

# 1. history CSV 로드
df = pd.read_csv(history_file)
print(df)

# 2. 에포크-Train Loss / R@K 시각화
plt.figure(figsize=(10, 6))

# Train Loss
plt.plot(df["epoch"], df["train_loss"], label="Train Loss", marker="o")

# Recall@K
plt.plot(df["epoch"], df["val_R@1"], label="Val R@1", marker="o")
plt.plot(df["epoch"], df["val_R@5"], label="Val R@5", marker="o")
plt.plot(df["epoch"], df["val_R@10"], label="Val R@10", marker="o")

plt.xlabel("Epoch")
plt.ylabel("Value")
plt.title(f"{backbone_name} Training History")
plt.grid(True)
plt.legend()
plt.show()
```
